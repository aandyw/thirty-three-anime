```python
import pandas as pd
import numpy as np
```

# Create Ratings List

```python
user_colnames = ['username', 'anime_id', 'my_score']
user_df = pd.read_csv('{}{}'.format(DATA_PATH, 'animelists_cleaned.csv'), usecols=user_colnames)
print("user_df: ", user_df.shape)
user_df.head()
```

```python
print(user_df.isnull().sum())
print('shape before drop:', user_df.shape)

user_df = user_df.dropna(axis=0)

print(user_df.isnull().sum())
print('shape after drop:', user_df.shape)
```

```python
user_df['username'] = user_df['username'].factorize()[0]
user_df['username'] += 1
user_df.rename(columns= {'username': 'uid'}, inplace=True)
user_df.rename(columns= {'anime_id': 'animeid'}, inplace=True)
user_df.rename(columns= {'my_score': 'rating'}, inplace=True)
user_df.head()
```

```python
user_df.to_csv('{}{}'.format(DATA_PATH, 'ratings.csv'), index=False, header=True)
```

# Create Animes List

```python
anime_colnames = ['anime_id', 'title', "type", "episodes", "rating", "score", "scored_by", "rank", "popularity", "favorites", "genre"]
anime_df = pd.read_csv('{}{}'.format(DATA_PATH, 'AnimeList.csv'), usecols=anime_colnames)
anime_df.head()

print(anime_df.isnull().sum())
print('shape before drop:', anime_df.shape)

anime_df = anime_df.dropna(axis=0)

print(anime_df.isnull().sum())
print('shape after drop:', anime_df.shape)
```

```python
anime_df.head()
```

```python
anime_df = anime_df.sort_values(by=['anime_id'])
anime_df.head()
```

## Creating animes.csv

```python
anime_df.rename(columns= {'anime_id': 'animeid'}, inplace=True)
anime_df.rename(columns= {'genre': 'genres'}, inplace=True)

anime_list = anime_df[['animeid', 'title', 'genres']]
anime_list.head()
```

```python
anime_list.to_csv('{}{}'.format(DATA_PATH, 'animes.csv'), index=False, header=True)
```

## Creating features.csv

```python
anime = anime_df
OHE_type = anime['type'].str.get_dummies() # one hot encoding
OHE_genre = anime['genres'].str.get_dummies(sep=',')
OHE_rating = anime['rating'].str.get_dummies()

anime.drop(['animeid', 'title', 'rating', 'type','genres'], axis=1, inplace=True)

# data = pd.concat([data, OHE_rating, OHE_type, OHE_genre], axis=1)
anime = pd.concat([anime, OHE_type, OHE_genre], axis=1)
anime.head()
```
